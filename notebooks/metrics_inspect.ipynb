```python
def prepare_df(df):
  # Create a new dataframe with the following properties: 
  # Rename "lgan_mmd_loss" to "MMD"
  # Rename "lgan_cov-CD" to "COV (%)" and normalize to 100
  # Rename "	1-NN-CD-acc" to "1-NNA (%)" and normalize to 100
  # Rename "fid" to "FPD"

  df = df.rename(columns={"lgan_mmd-CD": "MMD", "lgan_cov-CD": "COV (%)", "1-NN-CD-acc": "1-NNA (%)", "fid": "FPD"})
  df["COV (%)"] = df["COV (%)"] * 100
  df["1-NNA (%)"] = df["1-NNA (%)"] * 100

  # Groupby temperature and top_k
  df = df.groupby(["top_k", "temperature"]).mean()
  # Mask out everything that is not an important metric
  df = df.drop(columns=df.columns.difference(["MMD", "COV (%)", "1-NNA (%)", "FPD"]))


  return df

```

```python
# prepare train with z_score and export as latex
grouped_train = prepare_df(val_results_z_score)
grouped_train.to_latex('./submissions/report/results/train_metrics_z_score.tex', float_format="%.2f")
```
